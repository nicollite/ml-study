```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
inputs = pd.read_csv("entradas_breast.csv")
outputs = pd.read_csv("saidas_breast.csv")
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def create_network():
    # Create sequential classifier
    classifier = Sequential()

    units = round((inputs.columns.size + outputs.columns.size)/2)

    classifier.add(Dense(units = units, activation = 'relu', 
                     kernel_initializer = 'random_uniform',
                     input_dim = inputs.columns.size))
    classifier.add(Dense(units = units, 
                     activation = 'relu', 
                     kernel_initializer = 'random_uniform'))
    classifier.add(Dense(units = outputs.columns.size, activation = 'sigmoid'))
    # Set optimizer
    optimizer = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
    classifier.compile(optimizer = optimizer, 
                      loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn = create_network,
                            epochs = 100,
                            batch_size = 10)
```

```python
results = cross_val_score(estimator = classifier, X = inputs, y = outputs, cv = 10, scoring = "accuracy")
```
