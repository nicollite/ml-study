```python
import pandas as pd
```

```python
inputs = pd.read_csv("entradas_breast.csv")
outputs = pd.read_csv("saidas_breast.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train, inputs_tests, outputs_train, outputs_tests = train_test_split(inputs, outputs, test_size=0.25)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()

units = round((inputs.columns.size + outputs.columns.size)/2)

classifier.add(Dense(units = units, activation = 'relu', 
                     kernel_initializer = 'random_uniform',
                     input_dim = inputs.columns.size))
classifier.add(Dense(units = units, 
                     activation = 'relu', 
                     kernel_initializer = 'random_uniform'))
classifier.add(Dense(units = outputs.columns.size, activation = 'sigmoid'))
```

```python
optimizer = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
classifier.compile(optimizer = optimizer, 
                      loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
```

```python
classifier.fit(inputs_train, 
                  outputs_train,
                  batch_size = 10, 
                  epochs = 100)
```

```python
predics = classifier.predict(inputs_tests)
predics = (predics > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

precision = accuracy_score(outputs_tests, predics)
print(precision)

matrix = confusion_matrix(outputs_tests, predics)
print(matrix)
```

```python
result = classifier.evaluate(inputs_tests, outputs_tests)
print(result)
```
